```python
from constraint import Problem
problem = Problem()

# Variables
events = ["E1", "E2", "E3", "E4"]
speakers = ["S1", "S2", "S3", "S4"]
time_slots = ["T1", "T2", "T3", "T4"]

# domains
problem.addVariables(events, time_slots)
problem.addVariables(speakers, events)

for speaker in speakers:
    problem.addConstraint(
        lambda e1, e2, e3, e4: len(set([e1, e2, e3, e4])) == 4,
        ("E1", "E2", "E3", "E4"),
    )

problem.addConstraint(lambda t: t == "T1", ["E1"])
problem.addConstraint(lambda e: e == "E2", ["S2"])
problem.addConstraint(lambda t2, t3: t2 != t3, ("E2", "E3"))

solutions = problem.getSolutions()

# Print solutions
for solution in solutions:
    print(solution)
```

Problem-II

```python
from constraint import *

problem = Problem()

regions = [
    "Western Australia",
    "Northern Territory",
    "South Australia",
    "Queensland",
    "New South Wales",
    "Victoria",
    "Tasmania"
]
colors = ["Red", "Green", "Blue"]
problem.addVariables(regions, colors)

# Constraints
adjacent_regions = [
    ("Western Australia", "Northern Territory"),
    ("Western Australia", "South Australia"),
    ("Northern Territory", "South Australia"),
    ("Northern Territory", "Queensland"),
    ("South Australia", "Queensland"),
    ("South Australia", "New South Wales"),
    ("South Australia", "Victoria"),
    ("Queensland", "New South Wales"),
    ("New South Wales", "Victoria"),
]

for region1, region2 in adjacent_regions:
    problem.addConstraint(lambda r1, r2: r1 != r2, (region1, region2))
solutions = problem.getSolutions()

# Print solutions
print("\nAustralia Map Coloring Solutions:")
for solution in solutions:
    print(solution)
```
